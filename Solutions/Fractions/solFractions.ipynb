## [12 Punkte] Arithmetik

Implementieren Sie Methoden, um Arithmetik (Addition, Subtraktion, Multiplikation, Division) mit Brüchen auszuführen.

### Teilaufgaben

* **[2 Punkte]** Kehrwert: Implementieren Sie eine Methode mit dem Namen `invert()` zur Bildung des Kehrwertes eines Bruches. Diese gibt eine *neue* Instanz der Klasse zurück, bei welcher Zähler und Nenner im Vergleich zu der ursprünglichen Instanz vertauscht sind.

* **[3 Punkte]** Addition: Zwei Instanzen der `Fraction`-Klasse sollen sich addieren lassen. Das Resultat sei eine neue Instanz der Klasse, und die bestehenden werden nicht verändert. Dabei gilt in herkömmlicher Schreibweise für die Brüche `Fraction(a, b)` $= \frac{a}{b}$ und `Fraction(c, d)` = $\frac{c}{d}$: $$\frac{a}{b} + \frac{c}{d} = \frac{a \cdot d + c \cdot b}{b \cdot d}$$
- **[3 Punkte]** Subtraktion: Zwei Instanzen der `Fraction`-Klasse sollen sich subtrahieren lassen. Das Resultat sei eine neue Instanz der Klasse, und die bestehenden werden nicht verändert. Dabei gilt in herkömmlicher Schreibweise für die Brüche `Fraction(a, b)` $= \frac{a}{b}$ und `Fraction(c, d)` = $\frac{c}{d}$: $$\frac{a}{b} - \frac{c}{d} = \frac{a d - c b}{b d}$$

- **[2 Punkte]** Multiplikation: Zwei Instanzen der `Fraction`-Klasse sollen sich multiplizieren lassen. Das Resultat sei eine neue Instanz der Klasse, und die bestehenden werden nicht verändert. Dabei gilt in herkömmlicher Schreibweise für die Brüche `Fraction(a, b)` $= \frac{a}{b}$ und `Fraction(c, d)` = $\frac{c}{d}$: $$\frac{a}{b} \cdot \frac{c}{d} = \frac{a \cdot b}{c \cdot d}$$
 
- **[2 Punkte]** Division: Zwei Instanzen der `Fraction`-Klasse sollen sich dividieren lassen. Das Resultat sei eine neue Instanz der Klasse, und die bestehenden werden nicht verändert. Dabei gilt in herkömmlicher Schreibweise für die Brüche `Fraction(a, b)` $= \frac{a}{b}$ und `Fraction(c, d)` = $\frac{c}{d}$: $$\frac{a}{b} / \frac{c}{d} = \frac{a \cdot d}{b \cdot c}$$

## Ihre Lösung

````python
# This is a code gap. Students can fill it.
class Fraction:
    # Constructor
    def __init__(self, numerator, denominator=1):
        if denominator == 0:
            raise ValueError("Denominator can't be 0")
        self.numerator = numerator
        self.denominator = denominator

    # Representation
    # Repr function
    def __repr__(self):
        return f"Fraction({self.numerator},{self.denominator})"
    
    # String rep
    def __str__(self):
        return f'{self.numerator} / {self.denominator}'

    # Arithmetic
    # Invert    
    def invert(self):
        return Fraction(self.denominator, self.numerator)
    
    # Add
    def __add__(self, other):
        return Fraction(self.numerator * other.denominator + self.denominator * other.numerator, self.denominator * other.denominator)

    # Subtract
    def __sub__(self, other):
        return Fraction(self.numerator * other.denominator - self.denominator * other.numerator, self.denominator * other.denominator)

    # Multiply
    def __mul__(self, other):
        return Fraction(self.numerator * other.numerator, self.denominator * other.denominator)

    # Division
    def __truediv__(self, other):
        return Fraction(self.numerator * other.denominator, self.denominator * other.numerator)
    
    
    # Comparison
    # Greater
    def __gt__(self, other):
        return self.numerator * other.denominator > self.denominator * other.numerator

    # Less than
    def __ls__(self, other):
        return self.numerator * other.denominator < self.denominator * other.numerator

    # Greater or equals
    def __ge__(self, other):
        return not (self < other)

    # LEss or equals then
    def __le__(self, other):
        return not (self > other)
    
    # Equals
    def __eq__(self, other):
        return type(other) == Fraction and self.numerator == other.numerator and self.denominator == other.denominator
    # Unequal
    def __ne__(self, other):
        return not (self == other)
````
