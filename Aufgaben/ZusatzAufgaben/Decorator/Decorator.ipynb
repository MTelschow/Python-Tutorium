### Cache

Schreiben Sie einen Decorator mit dem Namen `cache`, der für jede Funktion ein Dictonary anlegt, dass die Eingabeparameter auf die Ausgabeparameter mappt und so, wenn das Ergebnis des Funktionsaufrufs bereits bekannt ist, einfach das Ergebnis ausgibt statt die Funktion aufzurufen. Um die Aufgabe zu vereinfachen, beschränken Sie sich auf Funktionen wie `fibonacci_rekursiv`, die lediglich ein positionelles Argument akzeptieren.

```python
# Schreiben Sie Ihre Antwort hier.
```

```python
@cache
def fibonacci_rekursiv(n):
    if n <= 1:
        return n
    return fibonacci_rekursiv(n - 2) + fibonacci_rekursiv(n - 1)

fibonacci_rekursiv(100)
```
