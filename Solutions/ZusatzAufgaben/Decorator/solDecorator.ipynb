### Überprüfe Eingabeparameter


Schreiben Sie einen Decorator mit dem Namen `check_string`, der überprüft, ob alle in die Funktion gegebenen Argumente aus Buchstaben bestehende strings sind. Wenn sich Ziffern oder Sonderzeichen im String befinden, soll der Decorator verhindern, dass die Funktion aufgerufen wird.


```python
# Schreiben Sie Ihre Antwort hier.
def check_string(func):

    def wrapper(*args, **kwargs):
        for arg in args:
            if not arg.isalpha():
                print("Argumente dürfen nur aus Buchstaben bestehen.")
                return

        for _, arg in kwargs.items():
            if not arg.isalpha():
                print("Argumente dürfen nur aus Buchstaben bestehen.")
                return

        return func(*args, **kwargs)

    return wrapper
```

```python
@check_string
def greet(name):
    print(f"Hello {name}!")


greet("Felix")
greet("X Æ A-12")
greet(name="Jan")
greet(name="Jane Doe")
```

```python
@check_string
def concat(*args):
    return " ".join(args)


print(concat("Merry", "Christmas"))
print(concat("Merry", "X-mas"))
```

### Cache


Schreiben Sie einen Decorator mit dem Namen `cache`, der für jede Funktion ein Dictonary anlegt, dass die Eingabeparameter auf die Ausgabeparameter mappt und so, wenn das Ergebnis des Funktionsaufrufs bereits bekannt ist, einfach das Ergebnis ausgibt statt die Funktion aufzurufen. Um die Aufgabe zu vereinfachen, beschränken Sie sich auf Funktionen wie `fibonacci_rekursiv`, die lediglich ein positionelles Argument akzeptieren.


```python
# Schreiben Sie Ihre Antwort hier.
def cache(func):
    cache = {}

    def wrapper(*args):
        if args not in cache:
            cache[args] = func(*args)
        return cache[args]

    return wrapper
```

```python
@cache
def fibonacci_rekursiv(n):
    if n <= 1:
        return n
    return fibonacci_rekursiv(n - 2) + fibonacci_rekursiv(n - 1)


fibonacci_rekursiv(100)
```
