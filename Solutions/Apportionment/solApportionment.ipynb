## [19 Punkte] Hare/Niemeyer-Verfahren

Da Sie nun die Stimmenverteilung als Dictionary haben, wollen Sie daraus die Sitzverteilung für ein Parlament bestimmen.

Das Hare/Niemeyer-Verfahren ist ein sehr einfaches Verfahren zur Sitzverteilung, dass auch ohne Computer, lediglich mit Stift und Papier, sehr schnell angewendet werden kann.
Aufgrund dieser Einfachheit genießt es große Beliebtheit, unter anderen bei den hessischen Landtagswahlen oder von 1987 bis 2005 bei der Bundestagswahl.
Da wir in dieser Klausur Ihre Programmierfähigkeit und nicht Ihre Rechenkünste testen wollen, sollen Sie dieses Verfahren in Python implementieren und nicht mit Stift und Papier lösen.

Im Hare/Niemeyer-Verfahren wird in einem einfachen Dreisatz die Anzahl der Stimmen pro Partei mit der Gesamtsitzanzahl multipliziert und dann durch die Gesamtstimmenzahl dividiert.
So wird ein idealer Sitzanspruch dieser Partei bestimmt.
Da jedoch nur ganze Sitze in einem Parlament vergeben werden können, werden diese Quotienten abgerundet.
Dadurch ist die Anzahl an verteilten Sitzen gleich oder geringer als die vorgegebene Gesamtsitzanzahl.

Die verbleibenden Sitze werden über die größten Reste im Quotienten vergeben. 
Wenn also zum Beispiel drei Sitze über den Rest verteilt werden, erhalten die drei Parteien mit dem größten Rest im Quotienten jeweils einen zusätzlichen Sitz.

Für den seltenen Fall, dass mehrere Parteien den gleichen Anspruch auf den/die letzten Sitz(e) haben, werden diese Sitze über ein Losverfahren vergeben.
Dieses Losverfahren wollen wir hier jedoch nicht implementieren.
Stattdessen soll lediglich ein Set ausgegeben werden, welche Parteien an einem Losverfahren teilnehmen würden.
Der Datentyp Set wurde hier gewählt, da die Reihenfolge der Einträge in diesem Fall keine Rolle spielen.

### Beispiel

Nehmen wir das Beispiel von oben:

---|---

Da so nur 7 Sitze vergeben wurde, müssen noch 3 weitere Sitze vergeben werden.
Dies richtet sich nach den größten Resten, in diesem Fall sind dies Gelb, Blau und Weiß mit 70.000, 61.000 und 60.000.
Daraus ergibt sich die folgende Sitzverteilung:

Partei | Stimmen | Ganzzahlquotient | Rest | Sitze

Für den Fall, dass nun 200 Wähler:innen ihre Stimme von Pink zu Rot wechseln, kommt es bei den Resten für den 3 Sitz zu einem Patt zwischen Weiß und Rot. In einem solchen Fall muss das ganze mit einem Los oder einer Stichwahl gelöst werden.
Für die Zwecke dieser Aufgabe wollen wir einfach die in Frage kommenden Parteien als Set ausgeben.

Partei | Stimmen | Ganzzahlquotient | Rest | Sitze
---|---|---|---|---
Gelb | 47.000 | 4 | **7.000** | 5
Weiß | 16.000 | 1 | ***6.000*** | 1 + Los
Rot | 16.000 | 1 | ***6.000*** | 1 + Los
Grün | 12.000 | 1 | 2.000 | 1
Blau | 6.100 | 0 | **6.100** | 1
Pink | 2.900 | 0 | 2.900 | 0
**Gesamt** | **100.000** | **7** | | **9 + Los**

Schreiben Sie nun die Funktion 'apportionment', mit der die Sitzverteilung berechnet werden kann. In jedem Fall muss Ihre Funktion die vorgegebene Signatur erfüllen. Ihre Funktion muss sich also mit einem Integer als erstes Argument und einem Dictionary als zweites Argument aufrufen lassen. Der Rückgabewert muss aus einem Dictionary und (einem Set bzw. None) bestehen.

```python
def apportionment(number_of_seats: int, votes: dict[str, int]) -> tuple[dict[str, int], set[str, ...] | None]:
    """ Apportion seats based of the Hare/Niemeyer-Method

    number_of_seats -- integer with the number of seats for apportionment
    votes -- dictionary mapping number of votes (integer) to party name (string)

    returns a dictionary mapping the number of seats (integer) to party name (string) and optionally a set of parties for the drawing procedure
    """
    # This is a code gap. Students can fill it.
    
    # Seats dict
    seats = dict()
    # Drawing of lots
    lots = set()
    # Total number of votes
    total_votes = 0
    # Seats remaining
    seats_free = number_of_seats
    # Rest after division
    rest = dict()
    
    # Get total votes
    for key in votes.keys():
        total_votes += votes[key]

    # Add Whole number quotient as votes
    for key in votes.keys():
        seats[key] = votes[key] * number_of_seats // total_votes # Whole number quotient
        
        seats_free -= seats[key] # Subtract the seats given
        
        rest[key] = votes[key] * number_of_seats % total_votes # Rest after divison

    # Add leftover votes and get lots
    while seats_free > 0:
        # Get highest rest
        max_rest = 0
        for key in rest.keys():
            if rest[key] > max_rest:
                max_rest = rest[key]

        # Get keys with highest rest into lots
        for key in rest.keys():
            if rest[key] == max_rest:
                lots.add(key)

        # If seats_free are not enough break out of loop
        if len(lots) > seats_free:
            break

        # Otherwise add seats to keys in lot and clear lot
        for key in lots:
            # Add seat
            seats[key] = seats[key] + 1 
            # Remove 1 seat left to assign
            seats_free -= 1
            # Remove rest
            rest[key] = 0
        # Clear
        lots.clear()
    
    
    return (seats, lots)
```
